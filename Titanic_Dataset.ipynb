```python
import pandas as pd
import numpy as np
```

```python
test_data = pd.read_csv('/content/test.csv')
train_data = pd.read_csv('/content/train.csv')
gender_data = pd.read_csv('/content/gender_submission.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_sex = encoder.fit_transform(train_data[['Sex']])
train_data = pd.concat([train_data, pd.DataFrame(encoded_sex, columns=['Sex_Male', 'Sex_Female'])], axis=1)
train_data.drop('Sex', axis=1, inplace=True)
```

```python
columns = ['Pclass', 'Age', 'Sex_Male', 'Sex_Female']
X = train_data[columns]
y = train_data['Survived']

X = X.dropna()
y = y.iloc[X.index]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 4)
```

```python
from sklearn.svm import SVC

model1 = SVC()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier()
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
```

```python
print(y_pred1)
print(y_pred2)
```

```python
from sklearn.metrics import accuracy_score

accuracy1 = accuracy_score(y_test, y_pred1)
accuracy2 = accuracy_score(y_test, y_pred2)
print("Accuracy SVM: ", accuracy1)
print("Accuracy Decision Tree: ", accuracy2)
```

```python
new_values = [[3, 61, 1, 0], [2, 50, 1, 0]]
predictionsSVM = model1.predict(new_values)
predictionsDT = model2.predict(new_values)

print(predictionsSVM)
print(predictionsDT)
```
